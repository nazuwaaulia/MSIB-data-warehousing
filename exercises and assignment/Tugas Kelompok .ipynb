# Nomor 3 (Python Part 2)
Buat Function yang menghitung faktorial dari sebuah bilangan tertentu.

```python
def tabel_perkalian(bilangan):
  for i in range (1,11):
    print (f"{bilangan} x {i} = {bilangan*i}")
```

```python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)

bilangan = int(input("Masukkan bilangan: "))
print(f"{bilangan}! = {factorial(bilangan)}")
```

# Nomor 4 (Python Part 2)
Buat Function untuk menghasilkan barisan Fibonacci hingga sejumlah bilangan tertentu yang ditentukan oleh pengguna.

```python
def fibonacci(n):
  fib_sequence = [0, 1]
  for i in range(2, n):
    fib_sequence.append(fib_sequence[i-1] + fib_sequence[i-2])
  return fib_sequence[:n]

n = int(input("Masukkan jumlah bilangan Fibonacci yang diinginkan: "))
print(fibonacci(n))
```

# Nomor 3 (Python Part 3)
Lakukan filtering untuk memperoleh baris-baris di mana nilai RM (jumlah kamar) lebih besar dari 6 dan MEDV (harga median rumah) lebih besar dari 25.
Berapa jumlah total baris yang memenuhi kriteria tersebut?


```python
import pandas as pd

url = 'https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv'
df = pd.read_csv(url, index_col=0, parse_dates=[0])
df.head()
```

```python
filtered_df = df.loc[(df['rm'] > 6) & (df['medv'] > 25)]

# Jumlah total baris yang memenuhi kriteria
total_rows = filtered_df.shape[0]
print(f"Jumlah total baris yang memenuhi kriteria: {total_rows}")
```
