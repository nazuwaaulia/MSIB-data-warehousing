# Data Aggregation

How do demographic factors (gender, senior citizen, partner, dependents) influence churn rates? Are certain demographic groups more likely to churn than others?

```python
telco_df = pd.read_csv('telco_customer_churn.csv')
```

```python
telco_df.info()
```

```python
telco_df.head(2)
```

```python
telco_df['Churn_Flag'] = telco_df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
telco_df.head(1)
```

```python
telco_df.groupby('gender',as_index=False)['Churn_Flag'].mean()
```

```python
telco_df.groupby('SeniorCitizen',as_index=False)['Churn_Flag'].mean()
```

```python
telco_df.groupby('Partner',as_index=False)['Churn_Flag'].mean()
```

```python
telco_df.groupby(['SeniorCitizen','Partner'],as_index=False)['Churn_Flag'].mean()
```

```python
telco_df.agg({'MonthlyCharges': ['mean', 'count']})
```

```python
telco_df.groupby('MonthlyCharges').aggregate(['mean', 'count'])
```

```python
telco_df['MonthlyCharges'].apply(['mean', 'count'])
```

```python
telco_df['MonthlyCharges'].summary(['mean', 'count'])
```
